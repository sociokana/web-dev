## Ex6-1. 日付を生成するイテレータ

クラス `year_date` のコンストラクタは西暦年を引数とし、
そのオブジェクトは、その年の日付を西暦年と月と日のタプルとして生成するイテレータとします。
クラス `year_date` を定義してください。西暦年が閏年かどうかは、
400で割り切れるか、100で割り切れずに4で割り切れるかで判定してください。

```
##########################################################
##  <[ ex6-1-year_date ]> 解答セル (Answer cell)
##  このコメントの書き変えを禁ず (Never edit this comment)
##########################################################

QUESTION_EXISTS = True # 質問がある場合は True に変えて，このセル内のコメントとして質問を記述してください
                        # Change to True if you have questions, and describe them as comments in this cell

## def __next__(self): の部分は下記のURL
## https://pydp.info/GoF_dp/behavior/16_Iterator/
## と教科書を参照しながらなんとなく作ったんですが、まだクラスのことが上手く理解できません。
## 模範解答のようなものは公開していただけるのでしょうか。


class year_date:
    
    def __init__(self, year):
        self.y = year
        self.month = [i+1 for i in range(12)]
        self.temp1 = [i+1 for i in range(30)]
        self.temp2 = [i+1 for i in range(31)]

        if self.y % 400 == 0:
            self.temp3 = [i+1 for i in range(29)]
        elif (self.y % 100 != 0) & (self.y % 4 == 0):
            self.temp3 = [i+1 for i in range(29)]        
        else:
            self.temp3 = [i+1 for i in range(28)]

        self.tuples = []
        self._i = 0

        if self.y % 400 == 0:
            for i in self.month:
                if i in [1, 3, 5, 7, 8, 10, 12]:
                    for j in self.temp2:
                        self.tuples.append((self.y, i, j))
                elif i == 2:
                    for j in self.temp3:
                        self.tuples.append((self.y, i, j))
                else:
                    for j in self.temp1:
                        self.tuples.append((self.y, i, j))

        elif (self.y % 100 != 0) & (self.y % 4 == 0):
            for i in self.month:
                if i in [1, 3, 5, 7, 8, 10, 12]:
                    for j in self.temp2:
                        self.tuples.append((self.y, i, j))
                elif i == 2:
                    for j in self.temp3:
                        self.tuples.append((self.y, i, j))
                else:
                    for j in self.temp1:
                        self.tuples.append((self.y, i, j))

        else:
            for i in self.month:
                if i in [1, 3, 5, 7, 8, 10, 12]:
                    for j in self.temp2:
                        self.tuples.append((self.y, i, j))
                elif i == 2:
                    for j in self.temp3:
                        self.tuples.append((self.y, i, j))
                else:
                    for j in self.temp1:
                        self.tuples.append((self.y, i, j))


    def __iter__(self):
        return self

    def __next__(self):
        if self._i == len(self.tuples):
            raise StopIteration()
        value = self.tuples[self._i]
        self._i += 1
        return value
```
