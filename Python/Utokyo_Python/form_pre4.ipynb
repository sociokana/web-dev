## Pre4-1. 校正

`filename` という名前のファイルを文字コード `code` で読み込み、
読み込んだテキストの中で、
`dic_proof` という辞書のキーとして登録されている文字列を、そのキーに対応する値に置き換えて、
その結果を `filename` の頭に `'new-'` を連結した名前のファイルに `code` で書き出す関数
`proof_text(filename, code, dic_proof)` を定義してください。

```
##########################################################
##  <[ pre4-1-proof_text ]> 解答セル (Answer cell)
##  このコメントの書き変えを禁ず (Never edit this comment)
##########################################################

QUESTION_EXISTS = False # 質問がある場合は True に変えて，このセル内のコメントとして質問を記述してください
                        # Change to True if you have questions, and describe them as comments in this cell

def proof_text(filename, code, dic_proof):
    with open(filename, 'r', encoding = code) as f, open('new-'+filename, 'w', encoding = code) as new:
        for line in f:
            for v, k in dic_proof.items():
                line = line.replace(v, k)
            new.write(line)
```

## Pre4-2. ディレクトリ中のファイル読み込み

このノートブックを実行しているColabの仮想環境に`'text'`という名前のディレクトリがあります。ディレクトリ `'text'` の中にあるファイルの名前 `filename` とそのエンコーディング `code` を受け取り、そのファイルの文字数を返す関数 `text_file_len()` を定義してください。

```
##########################################################
##  <[ pre4-2-text_file_len ]> 解答セル (Answer cell)
##  このコメントの書き変えを禁ず (Never edit this comment)
##########################################################

QUESTION_EXISTS = False # 質問がある場合は True に変えて，このセル内のコメントとして質問を記述してください
                        # Change to True if you have questions, and describe them as comments in this cell

def text_file_len(filename, code):
    length = int()

    with open('text/'+ filename, 'r', encoding = code) as f:
        for line in f:
            length += len(line)
    return length
```
